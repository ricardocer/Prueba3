## PRUEBA DE ALGUNAS FUNCIONES

```python
df_2 = df.loc[ :, ["FECHA_ACTUALIZACION","ID_REGISTRO","ENTIDAD_RES","MUNICIPIO_RES"] ]

cas_ent = list( df_2["ENTIDAD_RES"].values ) #Casos por entidad

# df_2.describe()

max_ent = max(df_2["ENTIDAD_RES"])
min_ent = min(df_2["ENTIDAD_RES"])

l_est = list( range(1, max_ent + 1 ) )
l_cas = [0]*( max_ent )

for i in range(0, max_ent):
    l_cas[i] = cas_ent.count(i+1)
```

```python
df_2
```

```python
import matplotlib.pyplot as plt

f, ax = plt.subplots( figsize=(15,8) )
plt.bar(l_est,l_cas)
plt.xlabel('Entidad Federativa')
plt.xticks(rotation=90)
plt.ylabel('Casos COVID-19')
plt.title('Casos COVID-19 por Entidad Federativa')
plt.show()
```

```python
prueba = dict ( zip(l_est,l_cas) )

df_3 = pd.DataFrame([[key, prueba[key]] for key in prueba.keys()], columns=['ENTIDAD_FEDERATIVA', 'NUMERO_CASOS'])

df_4 = df_3.sort_values("NUMERO_CASOS", ascending=[False] )

df_4.head()
```

```python
f, ax = plt.subplots( figsize=(18,32) )
plt.pie(df_4["NUMERO_CASOS"], labels = df_4["ENTIDAD_FEDERATIVA"], autopct = ("%1.1f%%") )
plt.title("CASOS COVID POR ENTIDAD_FEDERATIVA")
plt.show()
```

## Transformaciones de la base de datos

Se entiende por ***transformación*** a una modificación de la estructura de la base de datos, por ejemplo: añadir filas, quitar filas, agregar columnas, etc.

```python
df_5 = df.loc[ :, ["FECHA_ACTUALIZACION","ID_REGISTRO","ENTIDAD_RES","MUNICIPIO_RES"] ]

df_5["NUEVA_COLUMNA"] = 2*df_5["ENTIDAD_RES"]
#Se agregó una nueva columna, la cual es el doble de la columna "ENTIDAD_RES"

del( df_5["ENTIDAD_RES"] )
#Se elimina la columna "ENTIDAD_RES"

df_5.drop( ["FECHA_ACTUALIZACION", "NUEVA_COLUMNA"], axis=1)
#Se elimina una lista de columnas. En este caso, se eliminan: "FECHA_ACTUALIZACIÓN" y "NUEVA_COLUMNA"

df_5.drop( ["FECHA_ACTUALIZACION", "NUEVA_COLUMNA"], axis=1, inplace=True)
#Se elimina una lista de columnas. En este caso, se eliminan: "FECHA_ACTUALIZACIÓN" y "NUEVA_COLUMNA"
# inplace=True indica que se guardan las modificaciones en la base, sin necesidad de crear una nueva variable.

df_5.head()
```

## Groupby: obtener información esencial

Una de las funciones más interesantes que presenta el paquete **pandas** es **groupby**. 

En **groupby** se aplican operaciones matemáticas sencillas agrupando por categorías.

Se pueden aplicar diferentes funciones, quizá la más interesante sea: **.describe()**. Esta arroja la siguiente información:
* Cuántos casos totales hay
* Promedio
* Desviación Estándar
* Los 3 Cuartiles
* El máximo

Los **objetos groupby** pueden guardarse

```python
df.groupby(by="ENTIDAD_RES")["MUNICIPIO_RES"].max()
#Agrupa por ENTIDAD_RES y arroja el municipio con mayor número de contagios
# Mayor número de casos en cierto municio de cada ENTIDAD_RES
```

```python
df.groupby(by="ENTIDAD_RES")["MUNICIPIO_RES"].count()
#Esto agrupa los casos por ENTIDAD_RES
# Casos totales para cada ENTINDAD_RES
```

```python
df.groupby(by="ENTIDAD_RES")["MUNICIPIO_RES"].describe()
#Muestra un resumen estadístico agrupado para cada ENTIDAD_RES
```

Es posible aplicar un **groupby** a dos columnas y se hace de la siguiente manera:

**Observación:** Las columnas deben de ser el mismo tipo

```python
columnas = ["MUNICIPIO_RES","ENTIDAD_UM"]

df.groupby(by="ENTIDAD_RES")[columnas].max()
#Arroja el mayor número de contagios por MUNICIPIO_RES y ENTIDAD_UM, según la ENTIDAD_RES.
# No arroja cuál estado fue, solo arroja cuál es el mayor número para ALGÚN MUNICIPIO de dicha ENTIDAD_RES
```
